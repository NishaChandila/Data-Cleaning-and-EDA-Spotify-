```python
# importing libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/most-streamed-spotify-songs-2024/Most Streamed Spotify Songs 2024.csv',encoding='latin-1')
df.head(5)
```

### DATA CLEANING

```python
df.info()
```

```python
# Checking null values sum

df.isnull().sum()
```

```python
# creating df_missing that lists each column missing value
df_missing = pd.DataFrame(df.isnull().sum()).reset_index().rename(columns={'index': 'Track', 0: 'Missing Values'})

# drop first row 
df_missing.drop(index=df.index[0], axis=0, inplace=True)

# columns with missing values in descending order
df_missing[df_missing['Missing Values']>0].sort_values(by='Missing Values',ascending=False)
```

```python
# dropping 'tidal popularity' column

df.drop('TIDAL Popularity', axis=1, inplace=True)
df.info()
```

```python
# filling na values with 0 'spotify playlist count' column

df['Spotify Playlist Count'] = df['Spotify Playlist Count'].fillna(0)
df['Spotify Playlist Count'].isnull().sum()
```

```python
# replacing ',' from 'Spotify Playlist Count' column

df['Spotify Playlist Count'] = df['Spotify Playlist Count'].apply(lambda x:str(x).replace(',',''))
df['Spotify Playlist Count']
```

```python
# changing data type 'Spotify Playlist Count' column

df['Spotify Playlist Count'] = df['Spotify Playlist Count'].astype('int16')
df['Spotify Playlist Count'].dtype
```

```python
# identifing duplicates records

df_duplicates = pd.DataFrame(df.duplicated()).rename(columns={0: 'duplicate'})
df_duplicates[df_duplicates['duplicate']==True]
```

```python
df.iloc[2449:2451]
```

```python
# duplicate value sum

df.duplicated().sum()
```

```python
# identify outliers in 'Apple Music playlist count' and plotting box plot

plt.figure(figsize=(8,6))
df['Apple Music Playlist Count'].plot(kind='box')
plt.title('Boxplot of Apple Music Playlist Count')
plt.ylabel('Apple Music Playlist Count')
plt.show()
```

```python
# Correcting the date formatting

df['Release Date'] = pd.to_datetime(df['Release Date'])
```

```python
# extracting Month column from 'Release Date' column

df['Release Month'] = df['Release Date'].dt.month 
df['Release Date']
```

```python
df['Release Month']
```

```python
# extracting year column from 'Release date' column

df['Release Year'] = df['Release Date'].dt.year
df['Release Year']
```

```python
# Statiscal info for numerical values

df.describe()
```

```python
# 25th percentile for Track Score

q1 = np.percentile(df['Track Score'],25)
q1
```

```python
# 75th percentile for Track Score

q3 = np.percentile(df['Track Score'],75)
q3
```

```python
# finding IQR for Track Score

iqr = q3 - q1
iqr
```

```python
value_counts = df['Artist'].value_counts()
```

```python
value_counts.head(20)
```

```python
# Create the horizontal bar plot
plt.figure(figsize=(10,6))
value_counts.head(20).plot(kind='barh')

# Add labels and title
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.title('Frequency of Categories')
plt.gca().invert_yaxis()
plt.grid(axis='x')
plt.show()
```

```python
numerical_cols = [col for col in df.columns if df[col].dtype not in ['object','category','datetime64[ns]']]
numerical_cols
```

```python
correlation_matrix = df[numerical_cols].corr()

# Plotting the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create a scatter plot using 'Apple Music Playlist Count' and 'Deezer Playlist Count'

plt.figure(figsize=(10,6))
plt.scatter(df['Apple Music Playlist Count'], df['Deezer Playlist Count'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```
